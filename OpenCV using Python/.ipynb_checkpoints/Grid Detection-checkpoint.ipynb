```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('Images/chess.jpg')
img_show = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_show)
```

```python
found , corners = cv2.findChessboardCorners(img,(7,7))
found , corners
```

```python
cv2.drawChessboardCorners(img_show,(7,7),corners,found)
```
