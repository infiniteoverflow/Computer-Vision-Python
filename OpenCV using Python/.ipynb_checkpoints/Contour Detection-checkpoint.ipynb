```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('Images/contours.png',0)
image = image[50:450,100:300]
plt.imshow(image,cmap='gray')
```

```python
contour , heirarchy = cv2.findContours(image,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
heirarchy
```

```python
externals = np.zeros(image.shape)
```

```python
for i in range(len(contour)):
    
    if heirarchy[0][i][3] == -1:
        cv2.drawContours(externals,contour,i,255,-1)
```

```python
plt.imshow(externals,cmap='gray')
```

```python
internals = np.zeros(image.shape)
```

```python
for i in range(len(contour)):
    
    if heirarchy[0][i][3] != -1:
        cv2.drawContours(internals,contour,i,255,-1)
```

```python
plt.imshow(internals,cmap='gray')
```

```python
puppy = cv2.imread('Images/puppy1.jpg',0)
plt.imshow(puppy,cmap='gray')
```

```python
ret , thresh = cv2.threshold(puppy,170,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh,cmap='gray')
```

```python
contour , heirarchy = cv2.findContours(thresh,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
heirarchy
```

```python
externals = np.zeros(thresh.shape)
for i in range(len(contour)):
    if heirarchy[0][i][3] == -1:
        cv2.drawContours(externals,contour,i,255,-1)
plt.imshow(externals,cmap='gray')
```

```python
internals = np.zeros(thresh.shape)
for i in range(len(contour)):
    if heirarchy[0][i][3] != -1:
        cv2.drawContours(internals,contour,i,255,-1)
plt.imshow(internals,cmap='gray')
```
