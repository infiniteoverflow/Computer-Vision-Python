```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('Images/dogs.jpg')
image_show = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_show)
```

```python
template = image_show[180:400,660:880]
plt.imshow(template)
```

```python
methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR_NORMED','cv2.TM_CCORR','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    full_copy = image_show.copy()
    
    method = eval(m)
    
    res = cv2.matchTemplate(full_copy,template,method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
        
    height,width,channels = template.shape
    
    bottom_right = (top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(full_copy,top_left,bottom_right,color=(0,0,255),thickness=3)
    
    print(m)
    
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(121)
    ax.imshow(res)
    
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(121)
    ax.imshow(full_copy)
    
    plt.show()
    
```
