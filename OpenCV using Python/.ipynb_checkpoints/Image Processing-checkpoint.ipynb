## Color Mapping

RGB is a very old model. In 1970s , HSL (Hue,Saturation,Lightness) and HSV (Hue,Saturation,Value) models were introduced as an alternative for the RGB model. <br>

They were closely aligned with the way human perceived colors

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Opening an image

img = cv2.imread('Images/puppy1.jpg')
plt.imshow(img)
```

```python
# Converting the image to RGB fomat

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Converting RGB to HSV

img_hsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
```

```python
# Converting RGB to HSL

img_hsl = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img_hsl)
```

## Blending Images

To blend 2 images we use the formula : <br>

    new_pixel = α * pixel_img_1 + β * pixel_img_2 + γ

```python
img1 = cv2.imread('Images/puppy1.jpg')
plt.imshow(img1)
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread('Images/watermark.jpg')
plt.imshow(img2)
```

```python
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
# Resizing the image such that both are of same size

img1 = cv2.resize(img1,(1000,1000))
plt.imshow(img1)
```

```python
img2 = cv2.resize(img2,(1000,1000))
plt.imshow(img2)
```

```python
# Blending the two images

blended_img = cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.3,gamma=2)
plt.imshow(blended_img)
blended_img = cv2.cvtColor(blended_img,cv2.COLOR_BGR2RGB)
```

```python
cv2.imwrite('Images/blended.jpg',blended_img)
```

## Overlaying Images without blending

```python
img1 = cv2.imread('Images/puppy1.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread('Images/watermark.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
# Resizing img2 

img2 = cv2.resize(img2,(200,200))
plt.imshow(img2)
```

```python
larger_image = img1
smaller_img = img2
```

```python
larger_image[0:200,0:200] = smaller_img
plt.imshow(larger_image)
```
