```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Harris-Corner Method

```python
chess = cv2.imread('Images/chess.jpg')
chess_show = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
plt.imshow(chess_show)
```

```python
gray_image = cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray')
```

```python
gray_image = np.float32(gray_image)
```

```python
dst = cv2.cornerHarris(src=gray_image,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(chess)
```

```python
plt.imshow(chess_show)
```

## Shi-Thomasi Method

```python
chess = cv2.imread('Images/chess.jpg')
chess_show = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
plt.imshow(chess_show)
```

```python
corners = cv2.goodFeaturesToTrack(gray_image,100,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(chess,(x,y),3,(255,0,0),-1)

plt.imshow(chess)
```
