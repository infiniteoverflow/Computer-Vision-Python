```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creating a blank image

blank_img = np.zeros((512,512,3),dtype=np.int32)
```

```python
plt.imshow(blank_img)
```

<b> 

<b> Drawing a rectangle </b>

```python
cv2.rectangle(blank_img,pt1=(400,20),pt2=(500,80),color=(255,255,255),thickness=10)
plt.imshow(blank_img)
```

<b> Drawing a square at the center </b>

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness=10)
plt.imshow(blank_img)
```

<b> Drawing a Circle </b>

```python
cv2.circle(blank_img,center=(400,400),radius=70,color=(0,255,0),thickness=10)
plt.imshow(blank_img)
```

<b> Filling a shape </b>

```python
cv2.circle(blank_img,center=(400,400),radius=30,color=(255,255,255),thickness=-1)
plt.imshow(blank_img)

# Just change the thickness value to -1
```

<b> Draw a line </b>

```python
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(224,134,222),thickness=3)
plt.imshow(blank_img)
```

<b> Adding a text </b>

```python
font = cv2.FONT_ITALIC

cv2.putText(blank_img,text="Aswin",org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=2,lineType= cv2.LINE_4)
plt.imshow(blank_img)
```

<b> Drawing a custom Polygon </b>

```python
blank_img_2 = np.ones((512,512,3),dtype=np.int32)
plt.imshow(blank_img_2)
```

```python
vertices = np.array([[100,100],[200,200],[100,300],[300,200]])
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv2.polylines(blank_img_2,[pts],isClosed=True,color=(255,255,255),thickness=10)
plt.imshow(blank_img_2)
```
