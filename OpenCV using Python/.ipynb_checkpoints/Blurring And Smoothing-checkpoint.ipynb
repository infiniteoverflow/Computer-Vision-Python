```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_img():
    img = cv2.imread('Images/brick.jpeg').astype(np.float32) / 255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def disp_img(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
```

```python
disp_img(img)
```

### Gamma Correction

```python
gamma = 10
```

```python
res = np.power(img,gamma)
disp_img(res)
```

```python
gamma = 1/4
```

```python
res = np.power(img,gamma)
disp_img(res)
```

<b> If u decrease the value of gamma , the brightness of the image increases , while if u decrease the value of gamma , the brightness of the image decreases. </b>

```python
# Adding a text 

font = cv2.FONT_HERSHEY_COMPLEX

img = cv2.putText(img,text="Hello",org=(10,100),fontFace=font,fontScale=3,color=(255,0,0),thickness=1)

disp_img(img)
```
