```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
puppy = cv2.imread('Images/puppy_edge.jpg')
puppy_show = cv2.cvtColor(puppy,cv2.COLOR_BGR2RGB)
plt.imshow(puppy_show)
```

```python
puppy_gray = cv2.cvtColor(puppy,cv2.COLOR_BGR2GRAY)
plt.imshow(puppy_gray,cmap='gray')
```

```python
edges = cv2.Canny(image=puppy,threshold1=107,threshold2=127)
```

```python
plt.imshow(edges)
```

We need to reduce the noise

```python
# Bluring the image

blurred = cv2.GaussianBlur(puppy,ksize=(1,1),sigmaX=70)
```

```python
plt.imshow(blurred)
```

```python
median = np.median(puppy)
lower = int(max(0,0.7*median))
upper = int(min(255,1.2*median))
```

```python
edges = cv2.Canny(image=blurred,threshold1=lower+90,threshold2=upper+120)
plt.imshow(edges)
```
