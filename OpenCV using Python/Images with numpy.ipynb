```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
# Open the image
pic = Image.open('./Images/puppy1.jpg')
```

```python
pic
```

```python
# Converting to array

pic_arr = np.asarray(pic)
pic_arr
```

```python
# Plot the image array 

plt.imshow(pic_arr)
```

```python
pic_arr.shape
```

So the constraints of the image is:<br>
1. Height : 355 <br>
2. Width : 473 <br>
3. Color Channels : 3

<b>The three color channels include:</b>

```python
# Red Scale

plt.imshow(pic_arr[:,:,0])
```

```python
# Green Scale
plt.imshow(pic_arr[:,:,1])
```

```python
# Blue Scale
plt.imshow(pic_arr[:,:,2])
```

<b> The darker the color , lesser the percent of the particular scale color in the image </b>

```python
# Gray scale of Red Channel

plt.imshow(pic_arr[:,:,0],cmap="gray")
```

<b> To Completely remove a particular color channel from the pic , set that particular channel as 0 </b>

```python
pic_without_red = pic_arr.copy()
```

```python
plt.imshow(pic_without_red)
```

```python
# Removing the red scale

pic_without_red[:,:,0] = 0
```

```python
plt.imshow(pic_without_red)
```

```python
# Removing the green scale
pic_without_red[:,:,1] = 0
```

```python
plt.imshow(pic_without_red)
```

```python
# Adding a particular scale

pic_without_red[:,:,1] = 100
```

```python
plt.imshow(pic_without_red)
```
