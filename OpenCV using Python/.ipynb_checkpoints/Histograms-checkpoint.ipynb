```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/horse.jpeg')
img_cvt = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
```

```python
result = cv2.calcHist([img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
result
```

```python
plt.plot(result)
```

```python
result = cv2.calcHist([img_cvt],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(result)
```

```python
col = ('b','g','r')

for i,color in enumerate(col):
    res = cv2.calcHist([img_cvt],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(res)
    plt.xlim(0,50)
    plt.ylim(0,2500)

plt.title("Histogram of Horse")
```

```python
# Masking images and calculating Histogram for the given ROI

img = cv2.imread('Images/rainbow.jpg')
img_cvt = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
mask_cvt = cv2.cvtColor(mask,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(mask_cvt)
```

```python
mask[700:1000,600:1200] = 255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
mask_img = cv2.bitwise_and(img,img,mask=mask)
```

```python
plt.imshow(mask_img)
```

```python
hist_rainbow_mask = cv2.calcHist([img],channels=[1],mask=mask,histSize=[256],ranges=[0,256])
hist_rainbow = cv2.calcHist([img],channels=[1],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_rainbow_mask)
```

```python
plt.plot(hist_rainbow)
```

## Histogram Equalization

```python
gorilla = cv2.imread('Images/gorilla.jpeg',0)
gorilla_cvt = cv2.cvtColor(gorilla,cv2.COLOR_BGR2RGB)
plt.imshow(gorilla_cvt)
```

```python
hist_gorilla = cv2.calcHist([gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_gorilla)
```

```python
equi_gorilla = cv2.equalizeHist(gorilla)

plt.imshow(equi_gorilla,cmap='gray')
```

```python
# Equalization of Color Images

gorilla = cv2.imread('Images/gorilla.jpeg')
gorilla_cvt = cv2.cvtColor(gorilla,cv2.COLOR_BGR2RGB)
plt.imshow(gorilla_cvt)
```

```python
# Convert to HSV

gorilla_hsv = cv2.cvtColor(gorilla,cv2.COLOR_BGR2HSV)
```

```python
gorilla_hsv[:,:,2] # Takes only the V values in HSV
```

```python
plt.plot(cv2.calcHist([gorilla_hsv[:,:,2]],channels=[0],mask=None,histSize=[256],ranges=[0,256]))
```

```python
gorilla_hsv[:,:,2] = cv2.equalizeHist(gorilla_hsv[:,:,2])

plt.plot(cv2.calcHist([gorilla_hsv[:,:,2]],channels=[0],mask=None,histSize=[256],ranges=[0,256]))
```

```python
gorilla_cvt = cv2.cvtColor(gorilla_hsv,cv2.COLOR_HSV2RGB)
plt.imshow(gorilla_cvt)
```
