```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_img():
    img = cv2.imread('Images/brick.jpeg').astype(np.float32) / 255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def disp_img(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
```

```python
disp_img(img)
```

### Gamma Correction

```python
gamma = 10
```

```python
res = np.power(img,gamma)
disp_img(res)
```

```python
gamma = 1/4
```

```python
res = np.power(img,gamma)
disp_img(res)
```

<b> If u decrease the value of gamma , the brightness of the image increases , while if u decrease the value of gamma , the brightness of the image decreases. </b>

```python
# Adding a text 

font = cv2.FONT_HERSHEY_COMPLEX

img = cv2.putText(img,text="Hello",org=(10,100),fontFace=font,fontScale=3,color=(255,0,0),thickness=1)

disp_img(img)
```

```python
# Setting up the kernel for the low pass filter

kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
kernel
```

```python
# 2D filter

dst = cv2.filter2D(img,-1,kernel)
disp_img(dst)
```

```python
# Blurred

blurred = cv2.blur(img,ksize=(10,10))
disp_img(blurred)
```

```python
# Gaussian Blur

gauss = cv2.GaussianBlur(img,ksize=(7,7),sigmaX=1)
disp_img(gauss)
```

```python
noise = cv2.imread('Images/noise.jpg').astype(np.float32) / 255
noise = cv2.cvtColor(noise,cv2.COLOR_BGR2RGB)
disp_img(noise)
```

<b> 2D filters are used to reduce the noise in an image and to blur it </b>

```python
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
# 2D filter

dst = cv2.filter2D(noise,-1,kernel)
disp_img(dst)
```

<b> Here the noise has been removed from the image but at the cost of some details in the image </b>
