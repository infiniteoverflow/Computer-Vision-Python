```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/horse.jpeg')
img_cvt = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
```

```python
result = cv2.calcHist([img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
result
```

```python
plt.plot(result)
```

```python
result = cv2.calcHist([img_cvt],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(result)
```

```python
col = ('b','g','r')

for i,color in enumerate(col):
    res = cv2.calcHist([img],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(res)
    plt.xlim(0,50)
    plt.ylim(0,2500)

plt.title("Histogram of Horse")
```
