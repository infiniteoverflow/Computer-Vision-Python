```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Importing an image

img = cv2.imread('Images/puppy1.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
# Convert the color channel to its original one

fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# Adding the image as a GrayScale Image

img_gray = cv2.imread('Images/puppy1.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
# We need to convert from the default grayscale to original grayscale

plt.imshow(img_gray,cmap="gray")
```

```python
# Resizing an image

fix_img.shape
```

```python
new_image = cv2.resize(fix_img,(1000,1000))
plt.imshow(new_image)
```

```python
# Fliping an image

new_flip_h = cv2.flip(fix_img,0)
plt.imshow(new_flip_h)
```

```python
new_flip_v = cv2.flip(fix_img,1)
plt.imshow(new_flip_v)
```

```python
new_flip_both = cv2.flip(fix_img,-1)
plt.imshow(new_flip_both)
```

```python
# Saving an image 

cv2.imwrite('Images/puppy_cv2.jpg',img)
```
